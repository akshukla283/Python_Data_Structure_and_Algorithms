#### Create Stack class using list without size limit  


```python
class Stack:
    def __init__(self):
        self.list = []
        
    def __str__(self):
        
        values = [str(x) for x in self.list]
        values.reverse()
        return "\n".join(values)
    
    # checking stack
    def isEmpty(self):
        if self.list == []:
            return True
        else:
            return False
        
    # push 
    def push(self, value):
        self.list.append(value)
        return "Value has been successfully added to the Stack!"
    
    # pop
    def pop(self):
        if self.isEmpty():
            return "Stack is empty!"
        else:
            return self.list.pop()
    # peek 
    def peek(self):
        if self.isEmpty():
            return "Stack is empty!"
        else:
            return self.list[-1]
        
    # delete 
    def delete(self):
        self.list = None
        
        
        
tempStack = Stack()
tempStack.push(10)
tempStack.push(20)
tempStack.push(30)
tempStack.push(40)
tempStack.push(50)
print("Stack till last push : ")
print(tempStack)
print()
print("Pop operation :")
print(tempStack.pop())
print()
print("new push operation :")
tempStack.push(90)
tempStack.push(100)
tempStack.push(120)


print("Stack after new push : ")
print(tempStack)
print()

print("Peek value of Stack :")
print(tempStack.peek())
```

### Create Stack class using list with size limit

```python
class Stack:
    def __init__(self, maxSize):
        self.maxSize = maxSize
        self.list = []
        
        
    def __str__(self):
        values = [str(x) for x in self.list]
        values.reverse()
        return "\n".join(values)
    
    # check empty
    def isEmpty(self):
        if self.list == []:
            return True
        else:
            return False
        
    # check full
    def isFull(self):
        if len(self.list) == self.maxSize:
            return True
        else:
            False
    # push
    def push(self, value):
        if self.isFull():
            return "Stack is full!"
        else:
            self.list.append(value)
            return "Value has been added successfully to the Stack!"
        
    # pop
    def pop(self):
        if self.isEmpty():
            return "Stack is empty!"
        else:
            return self.list.pop()
    # peek
    def peek(self):
        if self.isEmpty():
            return "Stack is empty!"
        else:
            return self.list[-1]
        
    # delete
    def delete(self):
        self.list = None
        
tempStack = Stack(3)
tempStack.push(5)
tempStack.push(20)
tempStack.push(40)
print(tempStack.push(60))
print(tempStack)
print("pop operation : ")
print(tempStack.pop())
print("Stack after pop:")
print(tempStack)
print("peek value :")
print(tempStack.peek())

#print(tempStack.isEmpty())
            
    
```

### Create Stack using Linked List 

```python
class Node:
    def __init__(self, value = None):
        self.value  = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    # Iteration function
    def __iter__(self):
        node = self.head
        while node:
            yield node 
            node = node.next
            
class Stack :
    def __init__(self):
        self.LinkedList = LinkedList()
        
    def __str__(self):
        values = [str(x.value) for x in self.LinkedList]
        return "\n".join(values)
    
    # check empty
    def isEmpty(self):
        if self.LinkedList.head == None:
            return True
        else:
            return False
    # push 
    def push(self, value):
        new_node = Node(value)
        new_node.next = self.LinkedList.head
        self.LinkedList.head = new_node
        return "value has been added successfully to the Stack!"
    # pop
    def pop(self):
        if self.LinkedList.head is None:
            return "Stack is empty!"
        else:
            temp_value = self.LinkedList.head.value
            self.LinkedList.head = self.LinkedList.head.next
            return temp_value
        
    # peek 
    def peek(self):
        if self.LinkedList.head is None:
            return "Stack is empty!"
        else:
            temp_val = self.LinkedList.head.value
            return temp_val
    def delete(self):
        self.LinkedList.head = None
        
        
        
tempStack = Stack()
tempStack.push(1)
tempStack.push(2)
tempStack.push(3)
tempStack.push(5)
tempStack.push(10)
tempStack.push(20)
print("Stack till last push : ")
print(tempStack)
print("popped item :")
print(tempStack.pop())

print("Peek value ")
print(tempStack.peek())
#print(tempStack.isEmpty())
print("Values in Stack : ")
print(tempStack)
    
tempStack.delete()
print("After deleting Stack is : ")
print(tempStack)

        
```
